Binders, Paper, Phones etc are the most selled product and Supplies, Coiers etc are very less soled product so we have to find the reasons behind it

```python
%matplotlib inline
plt.figure(figsize=(20,15))
sns.barplot(x=df['Sub-Category'], y=df['Profit'])
```

The graph shows the profit/loss of different product in which copiers has most profit and tables gives most loss. But copies are the least sold so we have to increase its sales to maximise profit. Tables are not much sold and encounter a lot of loss so we have to either figure out the problem or we must drop this product

```python
states = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales', ascending=False)
plt.figure(figsize=(40,40))
states[:25].plot(kind='bar',edgecolor = "#000000", color=['red','yellow'])
plt.title("Profit/Loss & Sales of top 25 States")
plt.xlabel('States')
plt.ylabel('Sales and Net profit/loss')
plt.grid(True)
states[25:].plot(kind='bar', edgecolor="#000000", color=['red','yellow'])
plt.title("profit/loss & Sales of bottem 25 States")
plt.xlabel('States')
plt.ylabel('Sales and Net profit/loss')
plt.grid(True)
```

Tedas has a good amount of sales but it is encountring a loss it may be due to discount so we have to figure it out.  States like west Verginia, North Dakota has very less sales and discount so we ned to expend out market their.

```python
sub_cat = df.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales', ascending=False)
sub_cat.plot(kind='bar',color=['blue','orange'], edgecolor = "#000000")
```
